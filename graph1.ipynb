```python
import matplotlib.pyplot as plt
```

```python


# Sample data
x_values = df.Time
y1_values = df.AccV
y2_values = df.AccML
y3_values = df.AccAP

# Plotting
plt.plot(x_values, y1_values, color='red', label='AccV')
plt.plot(x_values, y2_values, color='green', label='AccML')
plt.plot(x_values, y3_values, color='blue', label='AccAP')

# Adding labels and title
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Line Graph with Multiple Y-columns')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x_values = df.Time
# y1_values = df.AccV
# y2_values = df.AccML
y3_values = df.AccAP
walking = df.Walking
turn = df.Turn
startHestation = df.StartHesitation

# Create a figure and axis
fig, ax1 = plt.subplots()

# Plotting numeric values on the first y-axis
ax1.plot(x_values, y1_values, color='red', label='AccV')
# ax1.plot(x_values, y2_values, color='green', label='AccML')
# ax1.plot(x_values, y3_values, color='blue', label='AccAP')
ax1.set_xlabel('X values')
ax1.set_ylabel('Numeric Values', color='black')
ax1.tick_params(axis='y', labelcolor='black')

# Create a second y-axis for action stages
ax2 = ax1.twinx()
ax2.plot(x_values, walking, color='orange', linestyle='--', label='Walking')
ax2.plot(x_values, turn, color='purple', linestyle='--', label='Turn')
ax2.plot(x_values, startHestation, color='pink', linestyle='--', label='StartHestation')
ax2.set_ylabel('Action Stages', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Adding title and legends
plt.title('Combined Line Graph with Numeric Values and Action Stages')
lines = ax1.get_lines() + ax2.get_lines()
labels = [line.get_label() for line in lines]
plt.legend(lines, labels, loc='upper left')

# Display the plot
plt.show()
```
